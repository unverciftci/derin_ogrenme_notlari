# Yapay Sinir Ağları ve Derin Öğrenme

Yapay sinir ağları (artifical neural networks), biyolojik sinir ağlarını model alarak geliştirilmiştir. Nöronlar bir sinyali alır, işlemden geçirir ve iletirler. Nöronlardan oluşan ağ, bizim vücudumuzu yönetir. 

neuron.jpg

Yapay sinir ağlarının her bir katmanı nöronlardan oluşur. Öğrenme görevinde, verinin arka planındaki yapıyı modellemek gerekir. Veri büyük ise katman ve nöron sayısı da ona göre artar. Dolayısıyla derin öğrenme ismi verilmişir.

artifial.jpg

Matematikte reel fonksiyonları temsil etmek veya onlara yaklaşım yapmak için genelde polinomları kullanıyoruz. Ancak çok değişkenli ve çok boyutlu fonksiyonları, mesela büyük verinin yapısını ifade etmek için yapay sinir ağları gereklidir. 

Makine öğrenmesinde doğrusal modeller, karar ağaçları, kümeleme yöntemleri gibi birçok etkili yöntem vardır. Ancak büyük ve yapılandırılmamış veride yapay sinir ağlarının üstünlüğü tartışmasızdır.
